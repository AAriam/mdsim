```python
from mdsim.initial_value_generator import Water
```

## Instantiation

```python
water_ensemble = Water(
    num_molecules_in_x=10,
    num_molecules_in_y=10,
    num_molecules_in_z=1,
    create_2d=True,
    temperature="310 K",
    pressure="1 atm",
    eq_bond_len="1.012 Å",
    eq_bond_angle="113.24 deg",
    unit_length="Å",
    unit_time="fs",
    random_seed=1111,
)
```

## Visualization

```python
water_ensemble.plot_ensemble()
```
